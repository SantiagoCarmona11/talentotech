```python
dia = 2
match dia:
    case 1:
        print("Lunes")
    case 2:
        print("Martes")
    case 3:
        print("Miercoles")
    case _:
        print("No es un dia de la semana")
```

```python
def sum(a, b):
    return (a + b)

a = int(input('Enter 1st number: '))
b = int(input('Enter 2nd number: '))

print(f'Sum of {a} and {b} is {sum(a, b)}')
```

```python
import datetime

def saluda_según_hora():
    hora_actual = datetime.datetime.now().hour

    if 6 <= hora_actual < 12:
        saludo = "Buenos días"
    elif 12 <= hora_actual < 18:
        saludo = "Buenas tardes"
    else:
        saludo = "Buenas noches"

    return saludo

print(saluda_según_hora())
```
