```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Variable independiente
y = np.array([2, 4, 5, 4, 5])                # Variable dependiente

# Crear el modelo de regresi√≥n lineal
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(x, y)

# Predecir valores
y_pred = modelo.predict(x)

# Mostrar resultados
print(f"Pendiente (coeficiente): {modelo.coef_[0]}")
print(f"Intersecci√≥n (ordenada al origen): {modelo.intercept_}")

# Graficar
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_pred, color='red', label='L√≠nea de regresi√≥n')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresi√≥n lineal simple')
plt.legend()
plt.show()
```
